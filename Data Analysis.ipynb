# TL Results Analysis

Analyse our transfer learning (TL) results with Python. The objective is to evaluate different TL methods' performance and to find out the statistical conclusion based on these numeric results.

## Load Required Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Set Environment

```python
root_path = '/Users/fangyijiewang/Desktop/PhD/Autumn_Year_1/'
```

```python
folder = 'numeric_data/'
```

## Load Data

```python
mobilenet_tr_loss = 'tr_100per_data.csv'
```

```python
mobilenet_tr_loss_df = pd.read_csv(root_path+folder+mobilenet_tr_loss)
mobilenet_tr_loss_df.head()
```

```python
mobilenet_tr_loss_df.shape
```

```python
mobilenet_tr_diffper_loss = 'tr_diff_per_data.csv'
```

```python
mobilenet_tr_diffper_loss_df = pd.read_csv(root_path+folder+mobilenet_tr_diffper_loss)
mobilenet_tr_diffper_loss_df.head()
```

## Results Visualization

```python
plt.plot(mobilenet_tr_loss_df)
plt.title('Training Loss Full Data', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.xticks(np.arange(0, 21, 5.0),fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Loss', fontsize=20)
plt.grid(True)
plt.legend(mobilenet_tr_loss_df.columns, loc ="upper right",fontsize=14,frameon=True)
#plt.legend('')
plt.show()
```

```python
plt.plot(mobilenet_tr_diffper_loss_df[['tr_222m_10per','tr_44m_10per','tr_unet_10per']])
plt.title('Training Loss 10 per Data', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.xticks(np.arange(0, 21, 5.0),fontsize=15)
plt.ylabel('Loss', fontsize=20)
plt.grid(True)
plt.legend(['tr_222m_10per','tr_44m_10per','tr_unet_10per'], loc ="upper right",fontsize=14)
#plt.legend('')
plt.show()
```

```python
plt.plot(mobilenet_tr_diffper_loss_df[['tr_222m_30per','tr_44m_30per','tr_unet_30per']])
plt.title('Training Loss 30 per Data', fontsize=20)
plt.xlabel('Epoch', fontsize=20)
plt.xticks(np.arange(0, 21, 5.0),fontsize=15)
plt.ylabel('Loss', fontsize=20)
plt.grid(True)
plt.legend(['tr_222m_30per','tr_44m_30per','tr_unet_30per'], loc ="upper right",fontsize=14)
#plt.legend('')
plt.show()
```
